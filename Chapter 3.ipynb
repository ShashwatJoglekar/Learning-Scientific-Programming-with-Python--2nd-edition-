### E3.2: The Sinc Function
We want to plot the function $$ f(x) = \frac{\sin(x)}{x} $$
At 0, this function should not be evaluated, so we will keep the value as f(0) = 1, as that's what the limit approaches. In reality, there should be no number there, but it will approach 1 so this solution is approximately fine

```python
import numpy as np
import matplotlib.pyplot as plt

x_values = np.linspace(-20, 20, 1000)
y_values = []

for x in x_values:
    if x == 0:
        y_values.append(1)
    else:
        y_values.append(np.sin(x) / x)

plt.plot(x_values, y_values, label='y = sin(x)/x', color='red')
plt.title('Plot of y = sin(x)/x')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
```

### E3.3: Some Oscillating Functions

### E3.4: Moore's Law
We will be creating an actual vs predicted plot for Moore's Law in logarithmic form:
$$ \log_{10}(n_i) = \log_{10}(n_0) + \frac{y_i - y_0}{T_2} \log_{10}2 $$

```python
import numpy as np
import matplotlib.pyplot as plt

t_2 = 2
y_0 = 1972
n_0 = 0.0025 * 10**6

## Actual data
year = [1972, 1974, 1978, 1982, 1985, 1989, 1993, 1997, 1999, 2000, 2003, 2004, 2007, 2008, 2012]
year = np.array(year)
ntrans = [0.0025, 0.005, 0.029, 0.12, 0.275, 1.18, 3.1, 7.5, 24.0, 42.0, 220.0, 592.0, 1720.0, 2046.0, 3100.0]
ntrans = np.array(ntrans) * 1.e6
y = np.log10(ntrans)

## Predicted
predicted = np.log10(n_0) + ((year - y_0) / t_2) * np.log10(2)

plt.plot(year, predicted, label='Predicted', color='blue')
plt.scatter(year, y, label='Actual Data', color='red')
plt.title('Predicted vs Actual Data of Transistor Counts')
plt.legend()
plt.xlabel('Year')
plt.ylabel('log10(Number of Transistors)')
```

### The Michaelis-Menten Equation
We want to plot v against [S] for the following equation:
$$ v = \frac{d[P]}{dt} = \frac{V_{max}[S]}{K_m + [S]}$$
Given the conditions that
$$ K_m = 0.04 M, V_{max} = 0.1Ms^{-1} $$

```python
import numpy as np
import matplotlib.pyplot as plt

V_max = 0.1
K_m = 0.04
S = np.linspace(0, 2, 1000)
v = (V_max * S) / (K_m + S)
plt.plot(S, v, label='Michaelis-Menten Kinetics', color='blue')
plt.title('Michaelis-Menten Kinetics')
plt.xlabel('Substrate Concentration (S)')
plt.ylabel('Reaction Velocity (v)')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
```

### P3.2.1: First Order Reaction Kinetics
We are looking at the kinetics of the reaction A -> B + C.
This solution won't go over how the equations are derived, but we get that
$$ [A] = [A]_0 e^{-(k_1 + k_2)t}$$
$$ [B] = \frac{k_1}{k_1 + k_2} [A]_0 (1 - e^{-(k_1 + k_2)t}) $$
$$ [C] = \frac{k_2}{k_1 + k_2} [A]_0 (1 - e^{-(k_1 + k_2)t}) $$
We want to plot this for
$$ k_1 = 300 s^{-1}, k_2 = 100 s^{-1}, [A]_0 = 2.0 mol dm^-3 $$

```python
import numpy as np
import matplotlib.pyplot as plt

a_0 = 2
k_1 = 300
k_2 = 100
t = np.linspace(0, 0.1, 1000)
a = a_0 * np.exp(-k_1 * t) + a_0 * np.exp(-k_2 * t)
b = (k_1 / (k_1 + k_2)) * (a_0 - a)
c = (k_2 / (k_1 + k_2)) * (a_0 - a)

plt.plot(t, a, label='[A](t)', color='blue')
plt.plot(t, b, label='[B](t)', color='orange')
plt.plot(t, c, label='[C](t)', color='green')
plt.title('Concentration vs Time for A -> B + C')
plt.xlabel('Time (s)')
plt.ylabel('Concentration')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.legend()
```

### P3.2.2: The Gaussian Prime Spiral
This is likely the most complicated problem in this chapter. A gaussian integer is a number whose real and imaginary parts are integers. A number x + iy is a Gaussian prime if: <br>
1) One of x and y is zero and the other is a prime number of the form 4n + 3 or -(4n + 3) for an integer n greater than or equal to 0 <br>
2) Both x and y are non zero and $$ x^2 + y^2 $$ is prime <br> 
Let us consider a sequence starting at some c_0 where an integer takes one step (+- 1) in its current direction but turns left if it encounters a gaussian prime. The initial direction is the positive real direction (Right). Let us write a spiral starting at $$ c_0 = 5 + 23i $$

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def is_gaussian_prime(number):
    if number[0] == 0:
        if number[1] % 4 == 3 and isprime(number[1]):
            return True
        else:
            return False
    elif number[1] == 0:
        if number[0] % 4 == 3 and isprime(number[0]):
            return True
        else:
            return False
    else:
        return isprime(number[0] ** 2 + number[1] ** 2)
    
def left_turn(vector):
    return (-vector[1], vector[0])

points = [[5, 23]]
direction = [1, 0]

for i in range(2000):
    last_point = points[-1]
    next_point = [last_point[0] + direction[0], last_point[1] + direction[1]]
    
    if is_gaussian_prime(next_point):
        direction = left_turn(direction)
    
    points.append(next_point)

re, im = zip(*points)
plt.plot(re, im, marker='o', markersize=2, linestyle='-', color='blue')
plt.title('Gaussian Primes Spiral')
plt.xlabel('Re')
plt.ylabel('Im')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.axis('equal')
```

### P3.3.2: The Lennard-Jones Interaction Potential
The Lennard-Jones Interaction Potential between two atoms at a distance r is given by:
$$ U(r) = \frac{B}{r^{12}} - \frac{A}{r^6} $$
Where A and B are constants. For argon they have the value:
A = 1.024e-23 Jnm^6 and B = 1.582e-26 Jnm^12. (We will be diving both by the Boltzmann constant for this code)
$$ k_b = 1.381e-23 JK^-1 $$
Part a)
Plot U(r) on one y-axis and on the second y-axis plot (both should be very large for small r):
$$ F(r) = \frac{-dU}{dr} = \frac{12B}{r^{13}} - \frac{6A}{r^7} $$ 
Part b)
For small displacements from interatomic equilibrium separation (F = 0), we can approximate the potential as a harmonic oscillator:
$$ V(r) = \frac{1}{2}k(r - r_0)^2 + \epsilon $$
Where:
$$ k = |\frac{d^2U}{dr^2}|_{r_0} = \frac{156B}{r_0^{14}} - \frac{42A}{r_0^8} $$

```python
import numpy as np
import matplotlib.pyplot as plt

## Part a

r = np.linspace(0.3, 0.8, 1000)
kb = 1.38e-23  # Boltzmann constant in J/K
a = 1.024e-23
b = 1.582e-26
a_0 = a / kb ## a used for plotting
b_0 = b / kb ## b used for plotting
interaction_potential = (b_0 / r**12) - (a_0 / r**6)
f_r = (12 * b_0)/(r ** 13) - (6 * a_0)/(r ** 7)

## Create a group of axes
fig, ax1 = plt.subplots()

ax1.plot(r, interaction_potential, label='Interaction_Potential', color='orange')
ax1.tick_params(axis='y', labelcolor='orange')

# Create a secondary y-axis
ax2 = ax1.twinx()
ax2.plot(r, f_r, label='Interatomic Force', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

ax1.set_ylim(-150, 100)  # Adjust limits for interaction potential
ax2.set_ylim(-1000, 1000)  # Adjust limits for interatomic force

ax1.legend(loc='lower right')
ax2.legend(loc='upper right')
plt.show()

## Part b
r_0 = (2 * b_0/a_0) ** (1/6)
epsilon = (b / r_0 ** 12) - (a / r_0 ** 6)
k = (156 * b)/(r_0 ** 14) - (42 * a)/(r_0 ** 8)
radii = np.linspace(0.3, 0.6, 1000)
interaction_potential = (b / radii**12) - (a / radii**6)
harmonic_approximation = ((1/2) * k * (radii - r_0) ** 2) + epsilon
plt.plot(radii, interaction_potential, label='Interaction Potential', color='orange')
plt.plot(radii, harmonic_approximation, label='Harmonic Approximation', color='blue')
plt.title('Interaction Potential and Harmonic Approximation')
plt.xlabel('Distance (nm)')
plt.ylabel('Potential Energy/K')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0.3, color='black', lw=0.5)
plt.legend()
plt.show()
```

### P3.3.3: The sunflower seed
For n seeds where s = 1, 2, ... n. We can place each seed a distance of $$r = \sqrt{s}$$ away from the origin at an angle $$ \theta = 2\pi s/\phi $$ where phi is some constant. For nature, that is the golden ratio:
$$ \phi = (1 + \sqrt{5}/2) $$

```python
import numpy as np
import matplotlib.pyplot as plt

golden_ratio = (1 + np.sqrt(5))/2
s = np.arange(1, 301, 1)
r = np.sqrt(s)
theta = (2 * np.pi * s) / golden_ratio

plt.polar(theta, r, label='Sunflower seed', color='green')
plt.title('Polar Plot of Sunflower Seed Arrangement', pad=20)  # Adjust the pad value to shift the title up
plt.legend()
```
